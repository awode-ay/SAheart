```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SAheart.csv",index_col=0)
print(df.shape[0]*df.shape[1])
print(df.columns)
df.sample(3)
```

```python
sns.countplot(df["chd"],hue=df["famhist"])
```

```python
df["age"].corr(df["sbp"])
```

```python
sns.distplot(df[df["chd"] == 1]["tobacco"],label="With CHD")
sns.distplot(df[df["chd"] == 0]["tobacco"],label="Without CHD")
plt.legend()
```

```python
sns.boxplot(x=df["chd"],y=df["tobacco"])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
def agegrp(age):
    if age <= 15:
        return "young"
    elif age > 15 and age <=35:
        return "adult"
    elif age > 35 and age <=55:
        return "mid"
    else:
        return "old"
df["age group"] = df["age"].apply(agegrp)
df.loc[:,["age","age group"]][0:10]
```

```python
sns.countplot(df["age group"],hue=df["chd"])
```

```python
sns.boxplot(x=df["age group"],y=df["ldl"])
```
